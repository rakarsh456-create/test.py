```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
```

## Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122th index, we can make a new column based on the Region

1 : "Begaia region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

Change the required columns as integer data type

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

Changing the other columns to float data type

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## let save the clean dataset 
df.to_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv')
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
## Monthly fire Analysis
dftemp=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Sidi-Bel Abbes Region",weight="bold")
```

```python
## Monthly fire Analysis
dftemp=df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Bejaia Region",weight="bold")
```

It is observed that August and September had the most number of firest fires for both regions.And from the above plot of months we can understand few things:

Most of the fires happened in August and very high fires happened only in 3 months - June, July and August

Less fires was on September
